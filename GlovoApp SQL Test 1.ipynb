### DDBB Creation

```python
import sqlite3 as lite
import random
import pandas as pd
```

```python
BBDD = 'GlovoAppEx1.db'
```

1 Creamos las tablas

```python
database = lite.connect(BBDD)
c = database.cursor()

try:
    c.execute('''
    DROP TABLE Orders
    ''')
    c.execute('''
    DROP TABLE Order_Points
    ''')
    
except:
    pass


query_create_orders = '''
CREATE TABLE "Orders" (
	"id"	INTEGER NOT NULL,
	"customer_id"	INTEGER NOT NULL,
	"courier_id"	INTEGER NOT NULL,
	"latitude_acceptance"	REAL NOT NULL,
	"longitude_acceptance"	INTEGER NOT NULL,
	PRIMARY KEY("id")
)
'''

query_create_orders_points = '''
CREATE TABLE "Order_Points" (
	"id"	INTEGER NOT NULL,
	"order_id"	INTEGER NOT NULL,
	"latitude"	REAL NOT NULL,
	"longitude"	REAL NOT NULL,
	"point_type"	TEXT,
	PRIMARY KEY("id")
)
'''


c.execute(query_create_orders)
c.execute(query_create_orders_points)

database.commit()


```

### DDBB Population

```python
def insert_order(order_id, customer_id, courier_id, latitude_acceptance, longitude_acceptance):
    database = lite.connect(BBDD)
    c = database.cursor()
        
    query_params = [order_id, customer_id, courier_id, latitude_acceptance, longitude_acceptance]
    
    query_insert_order = '''
    INSERT INTO Orders (id, customer_id, courier_id, latitude_acceptance, longitude_acceptance)
              VALUES(?,?,?,?,?)
    '''
    
    c.execute(query_insert_order, query_params)
    
    database.commit()
    

def insert_order_point(order_id, latitude, longitude, point_type):
    database = lite.connect(BBDD)
    c = database.cursor()
    
    query_params = [order_id, latitude, longitude, point_type]
    

    query_insert_order_point = '''
    INSERT INTO Order_Points (order_id ,latitude, longitude, point_type)
              VALUES(?,?,?,?)
    '''
    
    c.execute(query_insert_order_point, query_params)
    
    database.commit()
    
```

```python
def generate_order(id):
    order_id = id
    
    customer_id = random.randint(1, 100)
    courier_id = random.randint(1,100)
    
    latitude_acceptance = random.uniform(41.2,41.8)
    longitude_acceptance = random.uniform(2, 2.2)
    
    latitude_pick = random.uniform(41.2,41.8)
    longitude_pic = random.uniform(2, 2.2)
    
    latitude_del = random.uniform(41.2,41.8)
    longitude_del = random.uniform(2, 2.2)
    
    insert_order(order_id, customer_id, courier_id, latitude_acceptance, longitude_acceptance)
    insert_order_point(order_id, latitude_pick, longitude_pic, 'PICK UP')
    insert_order_point(order_id, latitude_del, longitude_del, 'DELIVERY')
    
    print(f'New Order Generated {order_id}')
```

```python
for order_id in range(1000):
    generate_order(order_id)
```

Function to calculate distance. Coding Haversine formula

```python
from math import sin, cos, sqrt, atan2, radians

# approximate radius of earth in km
def distance(lat1, lat2, lon1, lon2):
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    
    return distance
```

## Final Query



```python
SELECT 
	O.id as Order_Id,
	O.latitude_acceptance as CourierLat,
	O.longitude_acceptance as CourierLon,
	OPP.latitude as PickLat,
	OPP.longitude as PickLon,
	OPD.latitude as DelLat,
	OPD.longitude as DelLon
		

FROM Orders O
LEFT JOIN (Select * from Order_Points where point_type = 'DELIVERY') OPD
ON O.Id = OPD.order_id

LEFT JOIN (Select * from Order_Points where point_type = 'PICK UP') OPP
ON O.Id = OPP.order_id
```

```python
query_latslons = '''
SELECT 
	O.id as Order_Id,
	O.latitude_acceptance as CourierLat,
	O.longitude_acceptance as CourierLon,
	OPP.latitude as PickLat,
	OPP.longitude as PickLon,
	OPD.latitude as DelLat,
	OPD.longitude as DelLon
		

FROM Orders O
LEFT JOIN (Select * from Order_Points where point_type = 'DELIVERY') OPD
ON O.Id = OPD.order_id

LEFT JOIN (Select * from Order_Points where point_type = 'PICK UP') OPP
ON O.Id = OPP.order_id
'''
```

```python
import pandas as pd
import sqlite3

# Read sqlite query results into a pandas DataFrame
database = sqlite3.connect(BBDD)
df = pd.read_sql_query(query_latslons, database)

# Verify that result of SQL query is stored in the dataframe
print(df.head())

database.close()
```

```python
import numpy as np

df['Distance Courier-Pick'] = np.vectorize(distance)(df['CourierLat'], df['CourierLon'], df['PickLat'], df['PickLon'])
df['Distance Pick-Del'] = np.vectorize(distance)(df['PickLat'], df['PickLon'], df['DelLat'], df['DelLon'])
```
